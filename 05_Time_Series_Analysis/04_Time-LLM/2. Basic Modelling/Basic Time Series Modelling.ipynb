# Basic Time Series Modelling 

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.stattools import acf
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Stationarity 

## Autovariance function 

```python
# Simulate a purely random process
# from a general distribution
data = np.random.standard_normal(1000)
```

```python
print(len(data))
```

```python
acf_coef = acf(data)
```

```python
acf_coef
```

```python
plot_acf(data, lags=20);
```

## Randow walk 

```python
steps = np.random.standard_normal(1000)
steps[0]=0
random_walk = np.cumsum(steps)
```

```python
random_walk[:10]
```

```python
plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(random_walk)
plt.title("Simulated Random Walk")
plt.show()
```

```python
random_walk_acf_coef = acf(random_walk)
```

```python
plot_acf(random_walk, lags=20);
```

```python
random_walk_diff = np.diff(random_walk, n=1)
```

```python
plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(random_walk_diff)
plt.title('Noise')
plt.show()
```

```python
plot_acf(random_walk_diff, lags=20);
```

## Moving average 

```python
ar2 = np.array([2])
ma2 = np.array([1, 0.9, 0.3])

MA2_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)

plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(MA2_process)
plt.title('Moving Average Process of Order 2')
plt.show()

plot_acf(MA2_process, lags=20);
```
